##### Se puede observar que poseemos datos numericos, sin embargo, por medio de la exploracion de la informacion de los datos de las tablas, podemos observar que son de tipo object y deberian ser de tipo int64. Tambien podemos observar que tenemos datos faltantes. Estas inconsistencias se abordaran a continuación.

```python
# Se convierten los valores de la tabla "Natalidad" a tipo int64
Natalidad = Natalidad.astype('int64')
Natalidad.dtypes
```

```python
# Se abordan los datos faltantes en la tabla "Defunciones", los cuales se llenaran con la media de los datos de la columan objetivo
media_extranjero = Defunciones['Extranjero'].mean().astype('int64')
media_ignorado = Defunciones['Ignorado'].mean().astype('int64')

Defunciones['Extranjero'].fillna(value=media_extranjero, inplace=True)
Defunciones['Ignorado'].fillna(value=media_ignorado, inplace=True)

Defunciones.head()
```

##### Ahora se puede observar que no hay datos faltantes!
